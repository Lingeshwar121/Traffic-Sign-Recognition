```python
from skimage import feature
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import os    
import cv2

data_list = os.listdir("D:/Project/traffic_recognition/Project/Train")#display the file from the directory
print(data_list)

data = []
for x in data_list:
    temp_path = "D:/Project/traffic_recognition/Project/Train/" + x + "/" 
    fig_path = list(os.listdir(temp_path))
    for B_image in fig_path:
        path = "D:/Project/traffic_recognition/Project/Train/" + x + "/" 
        img = cv2.imread(os.path.join(path, B_image)) #load an image from specified file
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #color space conversion

        blob_doh = np.mean(feature.blob_doh(img))# Finds blobs in the given grayscale image.
        blob_log = np.mean(feature.blob_log(img)) #Finds blobs in the given grayscale image.
        canny = np.mean(feature.canny(img)) #Edge filter an image using the Canny algorithm
        corner_fast = np.mean(feature.corner_fast(img))#Extract FAST corners for a given image.
        corner_foerstner = np.mean(feature.corner_foerstner(img)) #Compute Foerstner corner measure response image.
        corner_harris = np.mean(feature.corner_harris(img))#Compute Harris corner measure response image
        corner_kitchen_rosenfeld = np.mean(feature.corner_kitchen_rosenfeld(img)#Compute Kitchen and Rosenfeld corner measure response image.
        corner_moravec = np.mean(feature.corner_moravec(img))#Compute Moravec corner measure response image
        corner_peaks = np.mean(feature.corner_peaks(img))#Find peaks in corner measure response image.
        corner_shi_tomasi = np.mean(feature.corner_shi_tomasi(img))#Compute Shi-Tomasi (Kanade-Tomasi) corner measure response image
        greycomatrix = np.mean(feature.greycomatrix(img, [2],[2]))#Calculate the gray-level co-occurrence matrix
        hessian_matrix = np.mean(feature.hessian_matrix(img))#Compute the Hessian matrix.
        hessian_matrix_det = np.mean(feature.hessian_matrix_det(img))#Compute the approximate Hessian Determinant over an image
        hessian_matrix_eigvals = np.mean(feature.hessian_matrix_eigvals(img))#Compute eigenvalues of Hessian matrix.
        local_binary_pattern = np.mean(feature.local_binary_pattern(img,2,2))#Gray scale and rotation invariant LBP (Local Binary Patterns).
        peak_local_max = np.mean(feature.peak_local_max(img))#Find peaks in an image as coordinate list or boolean mask.
        shape_index = np.mean(feature.shape_index(img))#Compute the shape index
        structure_tensor = np.mean(feature.structure_tensor(img))#Compute structure tensor using sum of squared differences.

        temp_img_features = [blob_doh, blob_log, canny, corner_fast, corner_foerstner, corner_harris, \
                             corner_kitchen_rosenfeld, corner_moravec, corner_peaks, corner_shi_tomasi, \
                             greycomatrix, hessian_matrix, hessian_matrix_det, hessian_matrix_eigvals, \
                             local_binary_pattern, peak_local_max, shape_index, structure_tensor, x
                            ]
        data.append(temp_img_features)
    print("Feature Extracted from folder : ", x)
        
```

```python
#Creating a dataframe using the extracted features
```

```python
df = pd.DataFrame(data)
df
```

```python
df.columns = ['blob_doh', 'blob_log', 'canny', 'corner_fast', 'corner_foerstner', 'corner_harris', \
                             'corner_kitchen_rosenfeld', 'corner_moravec', 'corner_peaks', 'corner_shi_tomasi', \
                             'greycomatrix', 'hessian_matrix', 'hessian_matrix_det', 'hessian_matrix_eigvals', \
                             'local_binary_pattern', 'peak_local_max', 'shape_index', 'structure_tensor', 'output'
                            ]
```

```python
df['output'] = pd.to_numeric(df['output'])
```
